```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#show charts in Jupyter notebook
%matplotlib inline

#adjust configuration of our plots
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#read in the data
df= pd.read_csv("movies.csv")
```

```python
#Look at first few columns
df.head()
```

```python
#Look for missing data using for loop

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))   #format our output

#result: there are null values in some of the columns
```

```python
#Check data types for our columns

df.dtypes
```

```python
#some current year values are incorrect-create a new column with correct year using released column
df["yearcorrect"] = df["released"].astype(str).str[0:4]
```

```python
#sort the dataframe by the gross column
df = df.sort_values(by=["gross"],inplace=False, ascending=False)
```

```python
#change workbook settings to view full df when we run it
pd.set_option("display.max_rows",None)
```

```python
#show all the unique values in the company column to check for data quality
df["company"].drop_duplicates().sort_values(ascending=False)
```

```python
#Hypothesis - budget will have a high correlation with gross column and certain companies
```

```python
#Test hypothesis with scatterplot-budget vs gross
plt.scatter(x=df["budget"],y=df["gross"])

plt.title("Budget vs Gross Earnings")

plt.xlabel("Gross Earnings")
plt.ylabel("Budget for Film")

#result: visually it seems to be correlated
```

```python
# create regression plot to plot budget vs gross

sns.regplot(x="budget",y="gross",data=df, scatter_kws={"color": "red"},line_kws={"color":"blue"})

#result: the line shows positive correlation
```

```python
#determine what the correlation is

df.corr(method="pearson") 

#result: high correlation between budget and gross
```

```python
#create a correlation matrix
correlation_matrix = df.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
#Look at company(non numeric)-so convert to have numeric representation of it to allow for correlation calculation
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype("category")  
        df_numerized[col_name] = df_numerized[col_name].cat.codes #gives category random numbers
        
df_numerized.head()
```

```python
#create correlation map with all numeric and converted non-numeric columns
correlation_matrix = df_numerized.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#view correlation in an easier way

sorted_pairs =  corr_pairs.sort_values()

sorted_pairs
```

```python
#show all pairs with high correlation-greater than 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr

#result: votes and budget have the highest correlation to gross earnings, company has low correlation to gross earnings
```
